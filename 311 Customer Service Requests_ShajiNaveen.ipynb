## Whether the average response time across complaint types is similar or not (overall)

#### H0:The Average Response time across Compliant types are similar
#### Ha: The Aeerage Response time Across Complaint types are not similar

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
mod = ols('Request_Closing_Time_Sec ~ C(Complaint_Type)', data = three11).fit()
```

```python
print(sm.stats.anova_lm(mod))
```

```python
three11.Descriptor.unique()
```

```python
## F value is not close to 1, which means the average response time is not similar.So Reject Null Hypothesis
```

### Are the type of complaint or service requested and location related?

#### H0: The complaint type and location is not associated
#### Ha: The complaint type and location is associated

```python
import scipy.stats as stats
from math import sqrt
```

```python
three11.rename({'Location Type' : 'Location_Type'}, axis = 1, inplace = True)
```

```python
contingency_table = pd.crosstab(three11.Complaint_Type, three11.Location_Type)
```

```python
contingency_table
```

```python
chisquare , p_value , df , expected_fre = stats.chi2_contingency(contingency_table)
```

```python
chisquare , p_value , df , expected_fre
```

```python
## The Chi Square Value is high, so reject Null hypothesis 
```
